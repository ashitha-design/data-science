```python
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
```

```python
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
```

```python
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null
```

```python
%%sql
SELECT DISTINCT LaunchSite
FROM SPACEXTBL;
```

```python
%%sql
SELECT * 
FROM SPACEXTBL
WHERE LaunchSite = "CCAFS LC-40"
LIMIT 5;
```

```python
%%sql
SELECT Orbit, COUNT(*) AS Count
FROM SPACEXTBL
GROUP BY Orbit;
```

```python
%%sql
SELECT Mission_Outcome, COUNT(*) AS Count
FROM SPACEXTBL
GROUP BY Mission_Outcome;
```

```python
%%sql
SELECT DISTINCT Booster_Version
FROM SPACEXTBL;
```

```python
%%sql
PRAGMA table_info(SPACEXTBL);
```

```python
%%sql
SELECT *
FROM SPACEXTBL
WHERE PAYLOAD_MASS__KG_ > 4000 
  AND PAYLOAD_MASS__KG_ < 6000;
```

```python
%%sql
SELECT Date, PAYLOAD_MASS__KG_
FROM SPACEXTBL
ORDER BY PAYLOAD_MASS__KG_ DESC
LIMIT 1;
```

```python
%%sql
SELECT *
FROM SPACEXTBL
WHERE Booster_Version LIKE "%FT%";
```

```python
%%sql
SELECT COUNT(*) AS SuccessfulDroneShipLandings
FROM SPACEXTBL
WHERE Landing_Outcome = "Success (drone ship)";
```

```python
%%sql
SELECT Landing_Outcome, COUNT(*) AS Count
FROM SPACEXTBL
WHERE Date BETWEEN "2010-06-04" AND "2017-03-20"
GROUP BY Landing_Outcome
ORDER BY Count DESC;
```
